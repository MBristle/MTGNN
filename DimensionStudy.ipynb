```python

import torch
from train_single_step import main
from util import get_single_default_args

args = get_single_default_args()

ds_name = 'rossman_2'
args.subgraph_size = 20
args.node_dim=40
args.batch_size = 2
args.epochs = 10

args.data = './data/rossman_cleaned_2.csv'
args.logger_name = 'new_ds_{}'.format(ds_name, args.horizon, args.num_nodes)

for num_nodes in [119]:
    args.horizon = 3
    args.num_nodes = num_nodes
    if num_nodes == 1:
        args.node_dim=1
        args.subgraph_size=1
    elif num_nodes <= 20:
        args.node_dim=num_nodes
        args.subgraph_size=num_nodes
    elif num_nodes < 40:
        args.node_dim=num_nodes
        args.subgraph_size=int(num_nodes / 2)
    args.save = "./model-{}-{}-{}.pt".format(ds_name, args.horizon, args.num_nodes)
    device = torch.device(args.device)
    print("Number of threads: {}".format(torch.get_num_threads()))
    print("Number of GCPU: {}".format(torch.cuda.device_count()))
    torch.set_num_threads(torch.get_num_threads())
    main(args, torch, device)
```
