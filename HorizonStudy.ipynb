```python

import torch
from train_single_step import main
from util import get_single_default_args

args = get_single_default_args()

ds_name = 'electricity'
args.subgraph_size = 20
args.batch_size =4
args.epochs=1

args.data ='./data/electricity.txt'
args.logger_name='experiment_horizon_{}'.format(ds_name,args.horizon,args.num_nodes)

for horizon in [2,4,8,16,32,64]:
    args.num_nodes =321
    args.horizon =horizon
    args.save="./model-{}-{}-{}.pt".format(ds_name,args.horizon,args.num_nodes)
    device = torch.device(args.device)
    print("Number of threads: {}".format(torch.get_num_threads()))
    print("Number of GCPU: {}".format(torch.cuda.device_count()))
    torch.set_num_threads(torch.get_num_threads())
    main(args,torch,device)
```
